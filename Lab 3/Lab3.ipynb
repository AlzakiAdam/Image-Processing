```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("Images/lena.tif")
```

```python
# Task 1.1
scale_factor = 2
new_size = (img.width * scale_factor, img.height * scale_factor)
img_scaled = img.resize(new_size, Image.LANCZOS)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_scaled, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Task 1.2
img_rotated = img.rotate(120, expand=True)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_rotated, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Task 1.3
shear = 0.2
w, h = img.size
new_w = int(w + abs(shear) * h)
img_shear = img.transform(
    (new_w, h),
    Image.AFFINE,
    (1, -shear, shear * h if shear > 0 else 0, 0, 1, 0),
    resample=Image.BILINEAR
)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_shear, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Task 2.1
img_arr = np.array(img, dtype=np.float64)
img_negative = Image.fromarray((255 - img_arr).astype(np.uint8))
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_negative, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Task 2.2
c = 255 / np.log(1 + img_arr.max())
img_log = Image.fromarray((c * np.log(1 + img_arr)).astype(np.uint8))
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_log, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Task 2.3
gamma = 0.1
c_pow = 255 / (255 ** gamma)
img_power = Image.fromarray((c_pow * (img_arr ** gamma)).astype(np.uint8))
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_power, cmap='gray')
plt.axis('off')
plt.show()
```
